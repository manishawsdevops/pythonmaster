```python
#Pre-Processing
# Rows with at least one NaN/null value
df[pd.isnull(df).any(axis=1)].head()
```

```python
# Drop the rows with the Nationality 'Chile'
df = df[df['Nationality'] != 'Chile']
df.reset_index() # reset's row indexes in case any rows were dropped
df.head()
```

```python
# Dropping multiple columns
df = df.drop(["Contract Valid Until","Height","Weight"], axis=1) # we won't use text features for our initial basic models
df.head()
```

```python
# Map feature values
df['Real Face'] = df['Real Face'].dropna().map( {'Yes': 1,'No': 0}).astype(int)
df.head()
```
